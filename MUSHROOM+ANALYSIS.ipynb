```python
data["ratio"]=data["weight"]/data["radius"]
```

```python
## Feature Scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data["radius"]=MinMaxScaler().fit_transform(data.radius.values.reshape(-1,1))
```

```python
data["ratio"]=MinMaxScaler().fit_transform(data.ratio.values.reshape(-1,1))
```

```python
data["weight"]=MinMaxScaler().fit_transform(data.weight.values.reshape(-1,1))
```

```python
dependent_variable="category"
```

```python
independent_variable=[x for x in data.columns if x not in [dependent_variable]]
```

```python
independent_variable
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
categorical_variable=data.dtypes[data.dtypes=="object"].index.values
```

```python
for var in categorical_variable:
    data[var]=le.fit_transform(data[var])
```

```python
### DecisionTreeClassifier using 5-Fold Cross Validation for tuning the parameter max_depth to avoid overfitting.
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,make_scorer
```

```python
params=[{'max_depth':[3,4,5,6,7,8,9,10]}]
```

```python
clf = GridSearchCV(DecisionTreeClassifier(),params, cv=5,
                       scoring=make_scorer(accuracy_score))
```

```python
clf.fit(data[independent_variable], data[dependent_variable])
```

```python
print(clf.best_params_)
```
